# Reviewing the Data

```python
#Now we can see how many columns we're working with.
#The null count will tell us which spaces we can expect missing data.
# i.e german/jap name only has 938/1028 non-null. meaning 90 have been lost in translation.
#From this info set, we can also infer that nearly half of the pokemon in this dex have dual typing.
dex.info()
```

```python
dex.describe()
```

```python
nulls_chart = pd.DataFrame(dex.isnull().any(), columns=['Nulls'])   
nulls_chart['# of_nulls [quantity]'] = pd.DataFrame(dex.isnull().sum())   
nulls_chart['# of_nulls [%]'] = round((dex.isnull().mean()*100),2)   
print(nulls_chart) 
```

# Data Manipulation

```python
#let's identify how many pokemon species we will be working with.

mons = str(len(dex))
first_string = 'There are '
second_string = ' Pokemon within the national dex.'
print(first_string + mons + second_string)
```

```python
#Let's organize the data. We don't really need the german and japanese names.
#Stats and effectiveness won't be what we need here either.
#We only need the essentials.
sort_dex = dex[['pokedex_number', 'name', 'generation', 'species', 'type_1', 'type_2']]
```

```python
#Here I want to remove the dex number and focus on the name so we can alphabetize the data.
sort_dex2 = dex[['name', 'generation', 'species', 'type_1', 'type_2']]
```

```python
# ascending Alphabetical order by Pokemon name A-Z
sort_dex2.sort_values('name')
```

```python
sort_dex2.sort_values(by = ['generation'], ascending = [True])
```

```python
#in order to sort by generation while maintaining alphabetical order, create dictionaries
#first remove the additional columns to avoid distractions

sort_dex3 = dex[['name', 'generation']]
sort_dex3
```

```python
sort_dex4 = dex[['name', 'generation','type_1','type_2','weight_kg','height_m']]
sort_dex4
```

```python
sort_dex5 = dex[['name', 'generation','type_1','type_2','hp','attack','defense','sp_attack','sp_defense','speed']]
sort_dex5
```

# Filtering Data & Conditionals

```python
sort_dex2.loc[(sort_dex2['generation']==3) & (sort_dex2['type_1']=='Rock')]
```

```python
sort_dex4.loc[(sort_dex4['generation']==4) & (sort_dex4['type_1']=='Rock') & (sort_dex4['weight_kg'] > 100)]
```

```python
sort_dex4.loc[(sort_dex4['type_1']=='Rock') & (sort_dex4['height_m'] >= 2) & (sort_dex4['weight_kg'] > 200)]
```

```python
#regular expressions/search with string
#sort out mega pokemon
import re

mega_list = sort_dex5.loc[sort_dex5['name'].str.contains('^Mega[a-z]*', flags=re.I, regex=True)]
mega_list
```

```python
mega_list = mega_list.drop(mega_list.index[16])
mega_list
```

```python
stats = ["hp", "attack", "defense", "sp_attack", "sp_defense", "speed"]
dex[stats].describe()
```

```python
for colm in stats:
    print('Max',colm,':',dex.loc[dex[colm].idxmax()]['name'])
    print('Min',colm,':',dex.loc[dex[colm].idxmin()]['name'])
```

# Visualization & Analysis

```python
gen_count = dex.groupby("generation").size().to_numpy()
plt.plot(np.arange(1,9),gen_count)
plt.xlabel('Generation')
plt.ylabel('Count')
print('Raw:',gen_count)
print('Mean:',gen_count.mean())
```

```python
ax = sns.countplot(x="type_1", data=dex);
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
ax.set(xlabel='Type 1', ylabel='Count', title='Pokemon Type');
```

```python
ax = sns.countplot(x="type_2", data=dex);
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
ax.set(xlabel='Type 2', ylabel='Count', title='Pokemon Type');
```
